```python
import pandas as pd

worksheet = gs.open("feedback response ").sheet1


data = worksheet.get_all_values()
print(data)
df = pd.DataFrame(data[1:], columns=data[0])
feedback_list = df['Feedback'].tolist()
print(feedback_list)

# model me data passing here
from transformers import pipeline

sentiment_pipeline = pipeline("sentiment-analysis")
for feedback in feedback_list:
  sentiment = sentiment_pipeline(feedback)[0]

  print(f"Feedback: {feedback}\nSentiment: {sentiment['label']} ({sentiment['score']:.2f})\n")
```
